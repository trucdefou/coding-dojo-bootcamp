Limpieza de datos

```python
#EN este caso es mejor eliminar los datos faltantes, ya que se tratan de datos de salud y el dataset no es muy extenso, correrir los datos con alguna tecnica de correccion podria modificar notablemente el resultado de cualquier analisis
df.drop(df[df['Cholesterol'] == 0].index, inplace = True)
df.drop(df[df['Cholesterol'] == 85].index, inplace = True)
df.drop(df[df['RestingBP'] == 0].index, inplace = True)
print(df['Cholesterol'].unique())
print(df['RestingBP'].unique()) # tablas sin valores invalidos
```

Obtencion de columna Factor de riesgo alto (Alta presion sistolica y alto colesterol)


```python
limite_RestingBP = 120 # se establece el limite para la columna RestingBP (mayor a 120 mmHg) se considera presion sistolica hipertensiva
df.loc[df['Cholesterol'] >= 240, 'Cholesterol_RF'] = 1
df.loc[df['Cholesterol'] < 240, 'Cholesterol_RF'] = 0 # se establecen los valores para la nueva columna Cholesterol_RF (Risk Factor)
limite_Cholesterol = 240 # se establece el limite para la columna Cholesterol (por encima de 240 mg/dL) se considera niveles altos
df.loc[df['RestingBP'] >= 120, 'RestingBP_RF'] = 1 # se establecen los valores para la nueva columna Resting_RF (Risk Factor)
df.loc[df['RestingBP'] < 120, 'RestingBP_RF'] = 0
df['Cholesterol_RF'] = df['Cholesterol_RF'].astype(int) # se convierten las columnas a valores enteros
df['RestingBP_RF'] = df['RestingBP_RF'].astype(int)
df['HighRiskFactor'] = df['RestingBP_RF'] + df['Cholesterol_RF'] #se realiza una suma y luego se evalua si hay alto factor de riesgo (si ambas situaciones de riesgo se presentan, se considera que existe alto factor de riesgo)
df.loc[df['HighRiskFactor'] >= 1, 'HighRiskFactor'] = 1
df.loc[df['HighRiskFactor'] < 0, 'HighRiskFactor'] = 0

print(df.head(10))

```

Se Generan valores relativos a los factores de riesgo por Colesterol alto y Presion sistolica en reposo. Si ambos son superiores a los valores normales (120mmHg y 240 mg/dL) se considera alto factor de riesgo.

A continuacion se generan agrupaciones por Edad y por Sexo. de casos de alto riesgo

```python
df2 = df.query('HighRiskFactor == 1') #se filtra por alto factor de riesgo
agrupados_edad = df2.groupby('Age') # se agrupa por edad
print(agrupados_edad.head(10))
colesterol_promedio_por_edad = agrupados_edad['Cholesterol'].mean().sort_values(ascending=False) #se halla el promedio de colesterol por edad
print(colesterol_promedio_por_edad)
agrupados_sexo = df2.groupby('Sex') # se agrupa por Sexo
print(agrupados_sexo.head(10))
colesterol_promedio_por_sexo = agrupados_sexo['Cholesterol'].mean().sort_values(ascending=False) #se halla el promedio de colesterol por sexo
print(colesterol_promedio_por_sexo)
def comorbilidad(x): #realizamos una funcion para generar una columna para indicar si poseen alguna enfermedad cardiaca
    if x < 1:
        return "No"
    else:
        return "Si"
df['Comorbilidad'] = df['HeartDisease'].apply(comorbilidad)
print(df.head())
```

Se Observa que mayor factor de riesgo con alto colesterol esta liderada por Personas de edad relativamente joven (32 anhos, seguidos de personas de 67, 40 y 65 anhos)
Tambien se observa que de las personas con factor de riesgo alto, las mujeres poseen mayores niveles de colesterol.

Se crea una columna para casos de comorbilidad para casos en donde exista la presencia de enfermedades cardiacas. se utiliza la funcion apply para ello.
